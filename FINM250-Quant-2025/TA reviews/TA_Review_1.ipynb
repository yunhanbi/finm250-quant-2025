### A Note on String Formatting

````python
# This is one of the most useful things in python, and you will be using
# it a lot both in this class, and in the future.

# {:.2f} means to print a float with 2 decimal places.
# {:,.2f} means to print a float with 2 decimal places and commas.
# {:,.2%} means to print a float with 2 decimal places, commas, and a percent sign (multiplied by 100).
# {:,d} means to print an integer.

print("The value of pi is {:.2f}.".format(3.141592653589793))
print("Wages went up by {:,.2%}.".format(0.05))
print("The population of the US is {:,d}.".format(327200000))

# We can also use f-strings.
print("-" * 80)
print(f"The value of pi is {3.141592653589793:.2f}.")

# Finally, alignment. You will probably not use this much.
print("-" * 80)
print(f'This is left aligned: {"Hello":<10}.')
print(f'This is right aligned: {"Hello":>10}.')
print(f'This is centered: {"Hello":^10}.')

# We can also use display() to show a variable.
x = 100
display(x)
````

#### What is an f-string?

An **f-string** (formatted string literal) is a way to embed expressions inside string literals, using curly braces `{}`. Introduced in Python 3.6, f-strings make string formatting easier and more readable.

**Syntax:**
- Start the string with an `f` or `F` before the opening quote: `f"..."` or `F'...'`.
- Place variables or expressions inside `{}` within the string.

**Example:**
```python
name = "Alice"
age = 30
print(f"My name is {name} and I am {age} years old.")
# Output: My name is Alice and I am 30 years old.
```

**You can also use formatting options:**
```python
pi = 3.14159
print(f"Pi rounded to 2 decimals: {pi:.2f}")
# Output: Pi rounded to 2 decimals: 3.14
```

**Comparison to `.format()`:**
- f-string: `f"Value: {value}"`
- `.format()`: `"Value: {}".format(value)`

**Advantages of f-strings:**
- More concise and readable
- Can include any valid Python expression inside `{}`
- Evaluated at runtime

**Summary:**
f-strings are the modern, recommended way to format strings in Python.

````python
import pandas as pd

hw1_data = pd.read_excel('multi_asset_etf_data.xlsx',
    sheet_name='excess returns',
    index_col=0,
    parse_dates=[0]
)
hw1_data.head()
````

````python
import numpy as np
# Build a dictionary to store the summary statistics, note that
# in pandas you *generally* want to avoid building DataFrames iteratively,
# ie. by appending rows and columns to a DataFrame one at a time, since
# there is a big performance hit.


def performance_summary(rets, adj_factor=12):
    # Here, we use the keyword argument adj_factor to adjust the
    # annualization factor, which, since in this course we will
    # mainly deal with monthly data, is set to 12 by default.

    summary = {}
    summary["Annualized Mean"] = rets.mean() * adj_factor
    summary["Annualized Volatility"] = rets.std() * np.sqrt(adj_factor)
    summary["Annualized Sharpe Ratio"] = (
        summary["Annualized Mean"] / summary["Annualized Volatility"]
    )
    return pd.DataFrame(summary, index=rets.columns)


# Since part b asks to find the best and worst Sharpe Ratios, we can
# use the .sort_values method to sort the DataFrame by the Sharpe Ratio.
metrics = performance_summary(hw1_data).sort_values(
    "Annualized Sharpe Ratio", ascending=False
)
metrics.style.format("{:.2}")
````

Why are we annualizing by $\sqrt{12}$? 

Assume that returns are independent identically distributed (i.i.d.), if we have monthly data, that means our monthly variance is going to be:
$$\sigma^2 = \frac{1}{n-1} \sum_{i=1}^{n} (r_i - \bar{r})^2$$

where $r_i$ is the return in month $i$, and $\bar{r}$ is the average return. And our annualized variance will be the *sum* of the monthly variances over the course of the year, which is $12 \sigma^2$.
Therefore, our annualized standard deviation will be $\sqrt{12 \sigma^2} = \sqrt{12} \sigma$.

What if my data isn't sampled equally? This comes up a lot in real life, where market data doesn't come in nice even intervals. In this case, we can use the log returns formula:

$$
\sigma^2 = \frac{1}{n-1} \sum_{i=1}^{n} \frac{\log(P_{i}/P_{i-1})^2}{t_i - t_{i-1}}
$$

The MV frontier. [link](https://bookdown.org/compfinezbook/introcompfinr/Efficient-portfolios-of.html)

````python
def tan_portfolio(mean_rets, cov_matrix):
    """
    Function to calculate tangency portfolio weights. Comes from the
    formula seen in class (Week 1).

    Args:
        mean_rets: Vector of mean returns.
        cov_matrix: Covariance matrix of returns.

    Returns:
        Vector of tangency portfolio weights.
    """
    inv_cov = np.linalg.inv(cov_matrix)
    ones = np.ones(mean_rets.shape)
    return (inv_cov @ mean_rets) / (ones.T @ inv_cov @ mean_rets)

def gmv_portfolio(cov_matrix):
    """
    Function to calculate the weights of the global minimum variance portfolio.

    Args:
        cov_matrix : Covariance matrix of returns.

    Returns:
        Vector of GMV portfolio weights.
    """
    try:
        cov_inv = np.linalg.inv(cov_matrix)
    except TypeError:
        cov_inv = np.linalg.inv(np.array(cov_matrix))

    one_vector = np.ones(len(cov_matrix.index))
    return cov_inv @ one_vector / (one_vector @ cov_inv @ (one_vector))


def mv_portfolio(mean_rets, cov_matrix, target=None, excess=True):
    """
    Function to calculate the weights of the mean-variance portfolio. If
    target is not specified, then the function will return the tangency portfolio.
    If target is specified, then we return the MV-efficient portfolio with the target
    return.
    
    **This assumes we have access to a risk-free asset**

    Args:
        mean_rets : Vector of mean returns.
        cov_matrix : Covariance matrix of returns.
        target (optional):  Target mean return. Defaults to None. Note: must be adjusted for
                            annualization the same time-frequency as the mean returns. If the
                            mean returns are monthly, the target must be monthly as well.
        excess (options): Whether or not we are using excess returns.

    Returns:
        Vector of MV portfolio weights.
    """
    w_tan = tan_portfolio(mean_rets, cov_matrix)

    if target is None:
        return w_tan
    
    elif not excess:
        w_gmv = gmv_portfolio(cov_matrix)
        delta = (target - mean_rets @ w_gmv) / (mean_rets @ w_tan - mean_rets @ w_gmv)
        return delta * w_tan + (1 - delta) * w_gmv
    else:
        ones = np.ones(mean_rets.shape)
        cov_inv = np.linalg.inv(cov_matrix)
        delta = (ones @ cov_inv @ mean_rets) / (mean_rets.T @ cov_inv @ ones) 
        return target * delta * w_tan



# Note: we are NOT annualizing here.
w_tan = mv_portfolio(hw1_data.mean(), hw1_data.cov())

w_tan_df = pd.DataFrame(w_tan, index=hw1_data.columns, columns=["Tangency Portfolio"])

# Here, we use the display() function to show both the weights and the performance summary,
# without having to make two separate cells, and not mess up the formatting caused by print().
display(
    w_tan_df.sort_values(by="Tangency Portfolio", ascending=False).style.format(
        "{:.2f}"
    )
)

# Calculate the portfolio returns.
w_tan_rets = hw1_data @ w_tan_df

# We could use linear algebra to directly calculate the
# portfolio statistics, but we will stick to using the
# performance_summary function for consistency, and also,
# if we want to calculate other risk metrics, we will need
# to use the performance_summary function.
tan_summ = performance_summary(w_tan_rets)
display(tan_summ.style.format("{:.2f}"))
````

Actually visualizing what the MV frontier looks like.

Intuitively: we have 2 portfolios, one of which has the optimal risk/return tradeoff, and the other of which has the lowest risk possible. The MV frontier will therefore be a combination of these two portfolios.

````python
# Plot the asset returns.
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('bmh')

plt.figure(figsize=(10, 6))

mean_rets = hw1_data.mean()
cov_matrix = hw1_data.cov()
w_tan = tan_portfolio(mean_rets, cov_matrix)
w_gmv = gmv_portfolio(cov_matrix)

for delta in np.linspace(-0.4, 0.4, 100):
    # Calculate the weights of the MV portfolio with the given delta.
    wts = delta * w_tan + (1 - delta) * w_gmv
    mu = mean_rets @ wts
    sigma = np.sqrt(wts.T @ cov_matrix @ wts)
    plt.scatter(sigma * np.sqrt(12), mu * 12, alpha=0.9, c='cornflowerblue', zorder=2)
    
# Add w_tan and w_gmv to the plot.
w_tan_mu = mean_rets @ w_tan
w_tan_sigma = np.sqrt(w_tan.T @ cov_matrix @ w_tan)
plt.scatter(
    w_tan_sigma * np.sqrt(12),
    w_tan_mu * 12,
    color="orange",
    label="Tangency Portfolio",
    zorder=3,
)
w_gmv_mu = mean_rets @ w_gmv
w_gmv_sigma = np.sqrt(w_gmv.T @ cov_matrix @ w_gmv)
plt.scatter(
    w_gmv_sigma * np.sqrt(12),
    w_gmv_mu * 12,
    color="green",
    label="Global Minimum Variance Portfolio",
    zorder=3,
)

    

plt.scatter(np.sqrt(np.diag(cov_matrix) * 12), mean_rets * 12, color="red", label="Assets", zorder=2)
plt.xlabel("Volatility (Standard Deviation)")
plt.ylabel("Mean Returns")
plt.title("Mean-Variance Frontier")
plt.tight_layout()
plt.show()
````
